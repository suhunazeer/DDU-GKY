```python
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('./titanic.csv')
data.head(10)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.isnull().mean()*100
```

```python
data.dropna(inplace=True)
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.info()
```

### Univariate Analysis

#### Histogram Plotting for "Survived column"

```python
sns.histplot(data=data, x='Survived')
plt.show()
```

#### Histogram plotting for Pclass column

```python
sns.histplot(data=data, x='Pclass')
plt.show()
```

#### Histogram plotting for Age

```python
sns.histplot(data=data, x='Age')
plt.show()
```

#### Histogram plotting for "Fare" column

```python
sns.histplot(data=data, x='Fare');
```

#### Box plotting for "Fare" column

```python
sns.boxplot(data=data, y='Fare');
```

#### KDE Plotting for "Survived column"

```python
sns.kdeplot(data=data.Survived)
plt.show()
```

```python
sns.kdeplot(data=data.Age)
plt.show()
```

#### Subplots

```python
sns.histplot(data=data)
plt.show()
```

```python
sns.boxplot(data=data)
plt.show()
```

##### subplots using for loops

```python
for i in data.columns:
    if data[i].dtypes !='object':
        sns.kdeplot(x=data[i])
        plt.show();
```

```python
age_upto_20=data.query(" Age<=20.0")
age_upto_20
```

```python
plt.figure(figsize=(10,6))
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title("Fare upto Age 20 - Bar Chart")
sns.barplot(data=age_upto_20,x='Age',y='Fare')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title("Fare upto Age 20 - line Chart")
sns.lineplot(data=age_upto_20,x='Age',y='Fare')
plt.show()
```

#### Count plot

```python
plt.figure(figsize=(4,3))
plt.xticks(rotation=90)
sns.countplot(data=data,x='Survived')
plt.show()
```
