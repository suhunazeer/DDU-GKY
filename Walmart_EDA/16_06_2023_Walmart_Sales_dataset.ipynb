### Feature Engineering Columns Celsius, Heat

### Temperature Conversion (-2&#176;F -32)*5/9

```python
data['Celsius']=[(Temp-32)*5/9 for Temp in data.Temperature]
```

```python
data.head()
```

```python
data['Heat']=pd.cut(x=data.Celsius, bins=[-20,10,35], labels=['Cold','Warm'])
```

```python
data.head()
```

```python
sns.barplot(data=data, x='Heat',y='Weekly_Sales', hue='Holiday_Flag')
plt.show()
```
