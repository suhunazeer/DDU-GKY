# <center><u> One Hot Encoding and Feature Scaling

```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

### Loading the dataset

```python
data = sns.load_dataset('titanic')
data.head()
```

```python
data.dropna(inplace = True)
data.drop_duplicates(inplace = True)
```

```python
data
```

#### One hot Encoding

```python
data['SEX'] = pd.get_dummies(data=data.sex, drop_first=True)
```

```python
data.head()
```

#### Converting the entire dataframe

```python
data_No_dummies = pd.get_dummies(data=data)
```

```python
data_No_dummies
```

```python
data_No_dummies.shape
```

```python
data_with_dummies = pd.get_dummies(data=data, drop_first=True)
```

```python
data_with_dummies
```

### importing MinMax Scaler, Standard Scaler, Robust Scalar, Label Encoder

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, LabelEncoder
```

#### <li>  <u> Label Encoder

```python
le = LabelEncoder()
```

```python
data['EMBARKED'] = le.fit_transform(data.embarked)
```

### Data Scaling or Data Normalisation

## <li> <u> MinMax Scaler

```python
minmax = MinMaxScaler()
```

```python
data[['age']]
```

```python
data[['age']].values
```

```python
age = data[['age']].values
```

```python
age
```

```python
data['AGE'] = minmax.fit_transform(age)
```

```python
data.head(10)
```

```python
data.AGE.min()
```

```python
data.AGE.max()
```

## <li> <u> Standard Scaler

```python
sc = StandardScaler()
```

```python
data['AGE_SC'] = sc.fit_transform(age)
```

```python
data.head(10)
```

```python
data.AGE_SC.min()
```

```python
data.AGE_SC.max()
```

```python
fare = data[['fare']].values
```

```python
data['FARE_SC'] = sc.fit_transform(fare)
```
