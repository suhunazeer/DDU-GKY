# <center><u> Scaling on Car Sale dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.get_dataset_names()
```

```python
data=pd.read_csv('./Car_sales.csv')
```

```python
data
```

```python
data.Manufacturer.unique()
```

```python
data.Model.unique()
```

```python
data.drop(columns={'Manufacturer','Model'},inplace=True)
```

```python
data
```

```python
data.dropna(inplace=True)
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

```python
data['VEH_TYPE'] = pd.get_dummies(data=data.Vehicle_type, drop_first=True)
```

```python
data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, LabelEncoder
```

### <li> Label Encoder

```python
le = LabelEncoder()
```

```python
data['VEHICLE_TYPE'] = le.fit_transform(data.Vehicle_type)
```

```python
data
```

### Data Scaling or Data Normalising

### <li><u> MinMax Scaler

```python
minmax = MinMaxScaler()
```

```python
data.Horsepower.max()
```

```python
data.Horsepower.min()
```

```python
data['HORSE_POWER'] = minmax.fit_transform(data[['Horsepower']].values)
```

```python
data.head()
```

```python
data.HORSE_POWER.min()
```

```python
data.HORSE_POWER.max()
```

### <li><u> Standard  Scaler

```python
sc = StandardScaler()
```

```python
data['HorsePower'] = sc.fit_transform(data[['Horsepower']].values)
```
