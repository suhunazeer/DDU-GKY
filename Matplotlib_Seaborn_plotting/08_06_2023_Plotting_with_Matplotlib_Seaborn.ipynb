## For loop to plot subplots

```python
for i in data.columns:
    if data[i].dtype !='object':
        sns.boxplot(y=data[i])
        plt.show()
```

```python
for i in data.columns:
    if data[i].dtype !='object':
        sns.histplot(x=data[i])
        plt.show()
```

```python
for i in data.columns:
    if data[i].dtype !='object':
        sns.kdeplot(data=data[i])
        plt.show()
```

## Subplots using enumerate

```python
features=[]
```

```python
for i in data.columns:
    if data[i].dtypes !='object':
        features.append(i)
```

```python
features
```

```python
list(enumerate(features))
```

```python
plt.figure(figsize=(13,10))
for i in enumerate(features):
    plt.subplot(5,3,i[0]+1)
    sns.boxplot(data=data, y=i[1], color='green')
```

```python
plt.figure(figsize=(13,10))
for i in enumerate(features):
    plt.subplot(5,3,i[0]+1)
    sns.kdeplot(data=data, x=i[1], color='brown')
```

```python
plt.figure(figsize=(13,10))
for i in enumerate(features):
    plt.subplot(5,3,i[0]+1)
    sns.histplot(data=data, x=i[1], color='orange')
```

## Line charts

```python
plt.figure(figsize=(10,6))
plt.xlabel('Year Resale Value')
plt.ylabel('Price')
plt.title('Line chart-Graphical EDA')
sns.lineplot(data=data, x='year_resale_value',y='Price_in_thousands', color='green')
plt.show()
```
